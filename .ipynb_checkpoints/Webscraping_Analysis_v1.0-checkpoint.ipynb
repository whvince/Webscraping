# <p><a name="Clean Data"></a></p>
### Clean Data

 - Import csv files and assign column names

```python
TA_reviews = pd.read_csv("tripadvisor_reviews_obao.csv", names = ['title','review', 'username', 'location', 'contribution', 'date_visit','date_review', 'rating'])
Yelp_reviews = pd.read_csv("yelp_reviews_obao.csv", names = ['review', 'username', 'location', 'date_review', 'rating'])
```

- Convert review date to date format

```python
Yelp_reviews.date_review = pd.to_datetime(Yelp_reviews['date_review'])
TA_reviews.date_review = pd.to_datetime(TA_reviews['date_review'])
```

- Re-index the data. The index will be later used for calculating the running average rating. 

```python
updated_Yelp_reviews = updated_Yelp_reviews.sort_values("date_review").reset_index(drop = True)
updated_TA_reviews = TA_reviews.sort_values("date_review").reset_index(drop = True)
```

# <p><a name="Analyze Data"></a></p>
## Analyze Data


- Compare the normalized rating distributions (in percentage) of two sites

```python
yelp = updated_Yelp_reviews.groupby("rating")[['review']].count()*100/updated_Yelp_reviews[['review']].count()
TA = updated_TA_reviews.groupby("rating")[['review']].count()*100/updated_TA_reviews[['review']].count()
yelp_TA_percentage = pd.merge(TA,Yelp,left_on = 'rating', right_on = 'rating', suffixes=('_Tripadvisor', '_Yelp'))

yelp_TA_percentage.plot(kind = "bar").yaxis.set_major_formatter(ticker.PercentFormatter())

plt.title('Tripadvisor has')
plt.xlabel('Rating')
plt.ylabel('Percentage')
plt.legend(["Tripadvisor", "Yelp"])
```

- Calculate and graph the running average of the ratings on Tripadvisor and Yelp
    - Tripadvisor rating has a surge at the end of 2019 and beginning of 2020

```python
updated_Yelp_reviews['running_avg'] = updated_Yelp_reviews.rating.cumsum()/(updated_Yelp_reviews.index+1)
updated_TA_reviews['running_avg'] = updated_TA_reviews.rating.cumsum()/(updated_TA_reviews.index+1)
```

```python
yelp_avg = updated_Yelp_reviews.groupby("date_review")[["running_avg"]].mean()
TA_avg = updated_TA_reviews.groupby("date_review")[["running_avg"]].mean()
yelp_TA_avg = pd.merge(yelp_avg, TA_avg, left_on = 'date_review', right_on = 'date_review', suffixes=('_Yelp','_Tripadvisor'))
yelp_TA_avg.plot(figsize=(20,6))

plt.title('Obao Rating on Tripadvisor Took a Positive Turn in Recent Months')
plt.xlabel('Date')
plt.ylabel('Running Average Rating')
```

- To investigate further on the pike of Tripadvisor average rating, first need to understand whether the number of reviews have increased for those two months. 


```python
# This is creating a monthly series (using end of the each month), which is helpful to calculate monthly number of reviews. 
updated_Yelp_reviews['date_monthly'] = updated_Yelp_reviews['date_review'] + pd.offsets.MonthEnd(0)
updated_TA_reviews['date_monthly'] = updated_TA_reviews['date_review'] + pd.offsets.MonthEnd(0)
```

```python
updated_TA_reviews.groupby("date_monthly")[["rating"]].count().plot.bar(figsize=(20,6))
plt.title("Number of Tripadvisor Reviews Surged in Recent Months")
plt.legend("")
plt.xlabel('Months')
plt.ylabel('Number of Reviews')
```

```python
updated_Yelp_reviews.groupby("date_monthly")[["rating"]].count().plot.bar(figsize=(30,6)) 
plt.title("Number of Yelp Reviews increased in November but came back to normal level since")
plt.legend("")
plt.xlabel('Months')
plt.ylabel('Number of Reviews')
```
